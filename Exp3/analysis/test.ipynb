```
AntiCo = pd.read_csv("../data/Coin.dat")
WOCo = pd.read_csv("../data/No_Coin.dat")
Channel = np.arange(1,1024)
```

```
plt.figure(figsize=(20,8))
plt.subplot(121)
plt.plot(Channel,AntiCo,label="With Anti-Coin.")
plt.plot(Channel,WOCo,label="W/O Anti-Coin.")
plt.xlabel("Channel")
plt.ylabel("Counts")
plt.legend()
plt.subplot(122)
plt.plot(Channel,AntiCo/AntiCo.sum(),label="With Anti-Coin., Norm")
plt.plot(Channel,WOCo/WOCo.sum(),label="W/O Anti-Coin., Norm")
plt.xlabel("Channel")
plt.ylabel("A.U.")
plt.legend()
plt.tight_layout()
plt.savefig("../plots/Spec.pdf")
```
